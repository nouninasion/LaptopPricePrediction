```python
import pandas as pd
import numpy as np
import re

def convert_to_gb(value):
    if 'TB' in value:
        return float(re.search(r'(\d+)', value).group(1)) * 1000
    elif 'GB' in value:
        return float(re.search(r'(\d+)', value).group(1))
    else:
        return 0

def extract_memory_parts(mem_string):
    mem_string = str(mem_string)
    parts = mem_string.split('+')
    
    ssd = 0
    hdd = 0

    for part in parts:
        part = part.strip()
        if 'SSD' in part:
            ssd += convert_to_gb(part)
        elif 'HDD' in part:
            hdd += convert_to_gb(part)
    return pd.Series([ssd, hdd])

train_data[['SSD', 'HDD']] = train_data['Memory'].apply(extract_memory_parts)
```
